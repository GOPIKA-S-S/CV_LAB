```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read the image
image = cv2.imread('/content/img1.jpg',cv2.IMREAD_GRAYSCALE)  # Read image as grayscale
print("Original image:")
cv2_imshow(image)
# Image negative
negative_image = 255 - image

# Display the negative image
print("Negative Image:")
cv2_imshow(negative_image)

```

Gray level Slicing

```python
# Assuming gray level slicing with a threshold range of 100 to 200
lower_threshold = 100
upper_threshold = 200

# Apply gray level slicing
output_image = np.where((image > lower_threshold) & (image < upper_threshold), 255, 0)

# Display the output image
cv2_imshow(output_image)

```

```python
# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display the equalized image
cv2_imshow(equalized_image)

```

```python
from matplotlib import pyplot as plt

# Perform Histogram Equalization
equalized_image = cv2.equalizeHist(image)

# Display the original image and its histogram
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.hist(image.ravel(), 256, [0, 256])
plt.title('Original Histogram')

# Display the equalized image and its histogram
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')

plt.subplot(1, 2, 2)
plt.hist(equalized_image.ravel(), 256, [0, 256])
plt.title('Equalized Histogram')

plt.show()
```
