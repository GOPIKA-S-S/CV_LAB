```python
import cv2
from google.colab.patches import cv2_imshow

# Load the image
img = cv2.imread('/content/img2.jpg')

```

Median Filtering

```python
# Apply median filtering
median = cv2.medianBlur(img, 5)  # Kernel size 5x5, you can adjust this value

# Display the original and filtered images
cv2_imshow(img)
cv2_imshow(median)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Max/Min Filtering:

```python

import numpy as np
# Apply max filtering
max_filtered = cv2.dilate(img, np.ones((5,5),np.uint8))  # Adjust the kernel size as needed

# Apply min filtering
min_filtered = cv2.erode(img, np.ones((5,5),np.uint8))  # Adjust the kernel size as needed

# Display the original, max filtered, and min filtered images
cv2_imshow(img)
cv2_imshow(max_filtered)
cv2_imshow(min_filtered)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Alpha-Trimmed Mean Filtering

```python
img = cv2.imread('/content/img2.jpg')

# Apply alpha-trimmed mean filtering
alpha_trimmed_mean = cv2.blur(img, (5, 5))  # Kernel size 5x5, you can adjust this value

# Display the original and filtered images
cv2_imshow(img)
cv2_imshow(alpha_trimmed_mean)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
